# Higgs Signal Classification Project

### **Objective:**
The goal of this project is to train a deep neural network (DNN) to separate the Higgs signal from different background processes using physics variables such as the Higgs mass and Z boson masses. After training, you will evaluate the performance of the model by plotting the following:

- **Receiver Operating Characteristic (ROC) Curve** and AUC to assess model performance.
- **Training and Validation Loss** to monitor model training progress.
- **DNN Score Distribution** to visualize how well the DNN separates signal from background.

---

### **Data:**
- **Signal:** Events corresponding to the Higgs boson decaying to four leptons (H → ZZ → 4l).
- **Background:** Various processes that mimic the Higgs signal but are not Higgs decays (e.g., ZZ → 4e, ZZ → 2e2mu, ZZ → 4mu).

#### **Key Variables:**
- `Higgs_mass`: Reconstructed mass of the Higgs candidate.
- `Z1_mass`: Mass of the first Z boson.
- `Z2_mass`: Mass of the second Z boson.

These variables are used to distinguish between signal and background.

---

### **Basic Project Steps**

1. **Data Preparation:**
   - Load both the signal and background datasets.
   - Extract the key variables (`Higgs_mass`, `Z1_mass`, `Z2_mass`) and create the target labels for signal and background.
   - Standardize the feature set using `StandardScaler` for better training performance.

2. **Splitting the Data:**
   - Split the dataset into training and testing sets.
   - Further split the training set into training and validation sets for monitoring the model’s performance during training.

3. **Model Construction:**
   - Build a simple feed-forward neural network using PyTorch with:
     - An input layer corresponding to the number of input features.
     - One hidden layer with a ReLU activation function.
     - An output layer with a sigmoid function for binary classification.

4. **Model Training:**
   - Use binary cross-entropy loss and the Adam optimizer to train the model.
   - Monitor both training and validation losses and implement early stopping if the validation loss stops improving.

5. **Evaluation:**
   - Plot the **ROC Curve** to visualize the tradeoff between true positive and false positive rates. Compute the **AUC** to summarize the performance for each background
   - Plot the **Training and Validation Loss** over time to check for overfitting or underfitting for each background
   - Plot the **DNN Score Distribution** to compare the model’s score for signal vs. background.
